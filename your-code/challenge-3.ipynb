## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
#sub1
quant_order=orders.groupby(["CustomerID"]).agg({"amount_spent" :"sum"})
 
#sub2                         
#quant_order["quantile_pos"] = pd.qcut(quant_order["amount_spent"], q=4, labels = False)# q =4 here is the quartile "given quantile"

#sub3. if we decide vip are at 0.95 quantile then we take vip at 19th ventile with q= 20

quant_order["quantile_pos"] = pd.qcut(quant_order["amount_spent"], q=20, labels = False)
quant_order[quant_order["quantile_pos"]>=19]
quant_order["Vip"] = quant_order["quantile_pos"]>=19
quant_order["prefered"] = (quant_order["quantile_pos"]>=15) & (quant_order["quantile_pos"]<19)

order2 = pd.merge(orders,quant_order, on="CustomerID")
order2
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
order2.groupby(["Vip","Country"]).agg({"amount_spent_y" :"sum", "Vip" : "count"})
order2["Vip"].value_counts()
order2[order2["Vip"]==True].groupby(["Country"]).agg({"amount_spent_y" :"sum", "Vip" : "count"})
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
# creating combined column 
order2["Combined"]= (order2["Vip"]== True) | (order2["prefered"] == True)
 
order2[order2["Combined"] == True ].groupby(["Country"]).agg({"Combined" : "count"})
```
